```python
def player(steep):
  if steep %2 ==1:
    return firstPlayer
  else:
    return secondPlayer

```

```python
def actions(game_board,player):
  actions = []
  for i in range(len(game_board)):
    for j in range(len(game_board[i])):
      if game_board[i][j] != 'X' and game_board[i][j] != 'O':
        action = [row[:] for row in game_board]
        action[i][j]=player
        actions.append(action)
  return actions
```

```python
def result(game_board,action):
  return action
```

```python
def terminal(game_board):
  if win_ai_player(game_board) or win_man_player(game_board):
    return True
  for i in game_board:
    for j in i:
      # print(j)
      if j != 'O' and j!='X':
        return False

  return True
```

```python
def win_ai_player(game_board):
  for i in range(len(game_board)):
    if game_board[i][0] == game_board[i][1] == game_board[i][2] == ai_player:
      return True
  for i in range(len(game_board)):
    if game_board[0][i] == game_board[1][i] == game_board[2][i] == ai_player:
      return True
  if game_board[0][0] == game_board[1][1] == game_board[2][2] == ai_player:
    return True
  if game_board[0][2] == game_board[1][1] == game_board[2][0] == ai_player:
    return True
  return False
```

```python
def win_man_player(game_board):
  for i in range(len(game_board)):
    if game_board[i][0] == game_board[i][1] == game_board[i][2] == man_player:
      return True
  for i in range(len(game_board)):
    if game_board[0][i] == game_board[1][i] == game_board[2][i] == man_player:
      return True
  if game_board[0][0] == game_board[1][1] == game_board[2][2] == man_player:
    return True
  if game_board[0][2] == game_board[1][1] == game_board[2][0] == man_player:
    return True
  return False
```

```python
#Minmax
```

```python
def utility(game_board):
  if win_ai_player(game_board):
    return 1
  elif win_man_player(game_board):
    return -1
  else:
    return 0
```

```python
def game_simulate(game_board, steep):
  # print_game_board(game_board)
  # print(steep)
  if terminal(game_board):
    return utility(game_board)

  current_player = player(steep)
  possible_scores = []

  for action in actions(game_board, current_player):
    next_game_state = result(game_board, action)
    score = game_simulate(next_game_state, steep + 1)
    possible_scores.append(score)
  return max(possible_scores) if current_player == ai_player else min(possible_scores)
```

```python
def get_optimal_steep(game_board, steep):


  current_player = player(steep)
  best_score = -float('inf')
  optimal_action_board = None

  for action_board in actions(game_board, current_player):
    score = game_simulate(action_board, steep + 1)
    if score > best_score:
      best_score = score
      optimal_action_board = action_board

  return optimal_action_board
```

```python
def cell_id_to_index(cell_id):
  switcher = {
    '1': [0,0],
    '2': [0,1],
    '3': [0,2],
    '4': [1,0],
    '5': [1,1],
    '6': [1,2],
    '7': [2,0],
    '8': [2,1],
    '9': [2,2],

  }
  return switcher.get(cell_id)
```

```python
def player_steep_valid(cell_choice):
  if cell_id_to_index(cell_choice) == None:
    return False
  if game_board[cell_id_to_index(cell_choice)[0]][cell_id_to_index(cell_choice)[1]] == 'X' or game_board[cell_id_to_index(cell_choice)[0]][cell_id_to_index(cell_choice)[1]] == 'O':
    return False
  return True
```

```python
def steepX(game_board, steep):

  current_player = player(steep)

  if current_player == man_player:
     cell_choice = input('Cell - ')
     while(player_steep_valid(cell_choice) == False):
       cell_choice = input('Invalid Cell - ')
     steep_index = cell_id_to_index(cell_choice)


     new_game_board = [row[:] for row in game_board]
     new_game_board[steep_index[0]][steep_index[1]] = man_player
     return new_game_board
  else:
    optimal_action_board = get_optimal_steep(game_board, steep)
    return optimal_action_board
```

```python
def print_game_board(game_board):
  print('-----------------')
  for i in game_board:
    print(*i)
  print('-----------------')
```

```python
game_board = [['1','2','3'],['4','5','6'],['7','8','9']]
ai_player = 'X'
man_player = 'O'
# firstPlayer = man_player
# secondPlayer = ai_player

firstPlayer = ai_player
secondPlayer = man_player


steep = 1

while True:

  print_game_board(game_board)


  if terminal(game_board):
    break

  game_board = steepX(game_board, steep)
  steep += 1

  if terminal(game_board):
    state = utility(game_board)
    print_game_board(game_board)
    if state == 1:
      print('AI Won!')
    elif state == -1:
      print('Man Won!')
    else:
      print('Draw!')
    break
```
